```python
import pandas as pd
```

```python
vehicle_df = pd.read_csv('../vehicles_us.csv')
```

```python
vehicle_df.head(20)
```

```python
vehicle_df.info()
```

```python
vehicle_df['is_4wd'] = vehicle_df['is_4wd'].astype('bool')
```

```python
vehicle_df.isna().sum() 
```

```python
vehicle_df.dropna(inplace=True)
```

```python
vehicle_df.duplicated().sum()
```

```python
vehicle_df['model_year'] = vehicle_df['model_year'].astype(int) 
```

```python
vehicle_df['odometer'] = vehicle_df['odometer'].astype(int) 
```

```python
vehicle_df['date_posted'] = pd.to_datetime(vehicle_df['date_posted'])   
```

```python
vehicle_df.info()
```

```python
vehicle_df.sample(20)
```

```python
import plotly.express as px
```

```python
#st.header('Price Distribution') 
#st.write('The price distribution of the vehicles is shown below.')
fig=px.histogram(vehicle_df, x='price', title='Price Distribution')
fig.show()
#st.plotly_chart(fig)
```

```python
#st.header('Vehicle model year vs price')
#st.write('The relationship between vehicle model year and price is shown below.')
fig=px.scatter(vehicle_df, x='model', y='price', title='Vehicle model vs price')
fig.show()
#st.plotly_chart(fig)    
```

```python
#st.header('Vehicle manufacturer vs odometer')
#st.write('The relationship between vehicle manufacturer and odometer is shown below.')
vehicle_df['manufacturer'] = vehicle_df['model'].str[:3]
fig=px.bar(vehicle_df, x='manufacturer', y='odometer', color='odometer', title='Vehicle manufacturer vs odometer')
fig.update_layout(yaxis=dict(range=[0,300000]))
fig.update_layout(
    plot_bgcolor='white'
)
fig.show()
#st.plotly_chart(fig)    
```

```python
fig = px.bar(
    vehicle_df,
    x='manufacturer', 
    y='odometer',
    color='odometer',             
    title='Vehicle manufacturer vs odometer'
)

fig.update_layout(
    yaxis=dict(range=[0, 300000]),
    plot_bgcolor='white'
)
fig.show()
```
